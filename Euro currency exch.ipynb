## Dropping the columns with more than 1000 empty rows

```python
x.drop(columns=['Slovak_koruna','Slovenian_tolar','Cypriot_pound','Estonian_kroon','Greek_drachma','Iceland_krona','Lithuanian_litas','Latvian_lats','Maltese_lira'], inplace = True)
```

```python
dtype_mapping = {
    'Korean_won': float,
    'Japanese_yen': float,
    'Indian_rupee': float,
    'Israeli_shekel': float,
    'Bulgarian_lev': float,
    'Indonesian_rupiah': float,
    'Hungarian_forint': float,
    'Croatian_kuna': float,
    'Hong_Kong_dollar': float,
    'UK_pound_sterling': float,
    'Danish_krone': float,
    'Czech_koruna': float,
    'Chinese_yuan_renminbi': float,
    'Swiss_franc': float,
    'Canadian_dollar': float,
    'Brazilian_real': float,
    'Australian_dollar': float
}

for col, dtype in dtype_mapping.items():
    x[col] = pd.to_numeric(x[col], errors='coerce').astype(dtype)
```

### Using the mean method of the observations to fill the null values

```python
x.fillna(x.mean(numeric_only=True).round(3), inplace=True)
```
